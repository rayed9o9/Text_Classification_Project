<h3>Importing libs and data</h3>

```python
import spacy
import sklearn.datasets as skd
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
import numpy as np
import nltk
```

```python
categories = ['family_friendly', 'horror']
data_train = skd.load_files('data/train', categories= categories, encoding= 'ISO-8859-1')
data_test = skd.load_files('data/test',categories= categories, encoding= 'ISO-8859-1')
```

```python
print(len(data_train.data) + len(data_test.data))
```

<h3>Data cleaning</h3>

<h3>Lowercasing</h3>

```python
for i in range(len(data_train.data)):
    data_train.data[i] = data_train.data[i].lower()

for i in range(len(data_test.data)):
    data_test.data[i] = data_test.data[i].lower()
```

<h3>special characters removing</h3>

```python
import re
for i in range(len(data_train.data)):
    data_train.data[i] = re.sub('[^A-Za-z]+', ' ', data_train.data[i])

for i in range(len(data_test.data)):
    data_test.data[i] = re.sub('[^A-Za-z]+', ' ', data_test.data[i])

data_train.data[5]
```

<h3>Remove Stope words</h3>

```python


from nltk.corpus import stopwords
# nltk.download('stopwords')
sw_nltk = stopwords.words("english")

def remove_stopwords(words):
    words = words.split()
    words_without_stopwords = ''
    for word in words:
        if word in sw_nltk:
            continue
        else:
            words_without_stopwords +=" "+ word
    return words_without_stopwords
```

```python
for i in range(len(data_train.data)):
    data_train.data[i] = remove_stopwords(data_train.data[i])

for i in range(len(data_test.data)):
    data_test.data[i] = remove_stopwords(data_test.data[i])

data_train.data[5]
```

```python
text_clf = Pipeline([('vect', CountVectorizer()), 
                      ('clf', MultinomialNB()) ])
```

```python
text_clf.fit(data_train.data, data_train.target)
```

```python
predicted = text_clf.predict(data_test.data)
```

```python
print('Accuracy achieved is ' + str(np.mean(predicted == data_test.target)))
print(metrics.classification_report(data_test.target, predicted, target_names=data_test.target_names)),
metrics.confusion_matrix(data_test.target, predicted)
```
